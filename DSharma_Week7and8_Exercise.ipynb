### Step 8 - Converting between string and datetime

```python
from datetime import datetime

```

```python
# lets learn about the column Timestamp of candy dataset.

print (df_candy16.Timestamp)
```

```python
df_candy16['Timestamp'] = pd.to_datetime(df_candy16['Timestamp'], format='%Y%m%d')
```

```python
# Convert datetype to string
df_candy16['Timestamp']=df_candy16['Timestamp'].astype(str)
```

```python
# lets see the datatype of column Timestamp now.

print (df_candy16.Timestamp)
```

```python
# Using to_datetime() & astype() to revert back the Timestamp column to datetime type.

df_candy16['Timestamp']=pd.to_datetime(df_candy16['Timestamp'].astype(str), format='%Y/%m/%d')
```

```python
# lets see the Timestamp column again
print (df_candy16.Timestamp)
```

### Step 9 - Convert timestamps to periods and back

```python
# Lets use to_period to change the timestamp column to period

df_candy16['Timestamp'] = df_candy16['Timestamp'].dt.to_period('M')
```

```python
# lets check the type for column Timestamp
print (df_candy16.Timestamp)
```

```python
# To convert back to timestamps, use to_timestamp

df_candy16['Timestamp'] = df_candy16['Timestamp'].dt.to_timestamp(how='end')
```

```python
# letc check the type of Timestamp column now.

print (df_candy16.Timestamp)
```
